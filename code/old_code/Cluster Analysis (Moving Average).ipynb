```python
#Find product and sum for each cluster
clusters_product2 = {}
clusters_sum2 = {}

for half in h_centuries2:
    clusters_product2[half] = clusters.copy()
    p = []
    for ind,row in clusters_product2[half].iterrows():
        t1t2 = str(int(row['Topic 1'])) + ',' + str(int(row['Topic 2']))
        t1t3 = str(int(row['Topic 1'])) + ',' + str(int(row['Topic 3']))
        t2t3 = str(int(row['Topic 2'])) + ',' + str(int(row['Topic 3']))
        share_product = ctopics2[half][t1t2] * ctopics2[half][t1t3] * ctopics2[half][t2t3]
        p.append(share_product)
    clusters_product2[half]['Product'] = p
    clusters_product2[half].sort_values('Product', ascending=False, inplace=True)
    
for half in h_centuries2:
    clusters_sum2[half] = clusters.copy()
    s = []
    for ind,row in clusters_sum2[half].iterrows():
        t1t2 = str(int(row['Topic 1'])) + ',' + str(int(row['Topic 2']))
        t1t3 = str(int(row['Topic 1'])) + ',' + str(int(row['Topic 3']))
        t2t3 = str(int(row['Topic 2'])) + ',' + str(int(row['Topic 3']))
        share_sum = ctopics2[half][t1t2] + ctopics2[half][t1t3] + ctopics2[half][t2t3]
        s.append(share_sum)
    clusters_sum2[half]['Sum'] = s
    clusters_sum2[half].sort_values('Sum', ascending=False, inplace=True)
```

```python
#Select corners for each half century using a product
clusters_product3 = {}

for half in h_centuries2:
    group = [5,9,10,18,22,26,32,35,46,50,55,60,2,19,20,15,54,56]
    clusters_product3[half] = pd.DataFrame(columns = ['Topic 1', 'Topic 2', 'Topic 3', 'Product'])
    for ind,row in clusters_product2[half].iterrows():
        if (row['Topic 1'] in group) or (row['Topic 2'] in group) or (row['Topic 3'] in group):
            pass
        else:
            clusters_product3[half] = clusters_product3[half].append(row, ignore_index=True)
            group.append(row['Topic 1'])
            group.append(row['Topic 2'])
            group.append(row['Topic 3'])

    line1 = []
    line2 = []
    line3 = []
    for ind,row in clusters_product3[half].iterrows():
        line1.append(words[int(row['Topic 1'])-1])
        line2.append(words[int(row['Topic 2'])-1])
        line3.append(words[int(row['Topic 3'])-1])

    clusters_product3[half]['Words 1'] = line1
    clusters_product3[half]['Words 2'] = line2
    clusters_product3[half]['Words 3'] = line3
```

```python
#Repeat with sum
clusters_sum3 = {}

for half in h_centuries2:
    group = [5,9,10,18,22,26,32,35,46,50,55,60,2,19,20,15,54,56]
    clusters_sum3[half] = pd.DataFrame(columns = ['Topic 1', 'Topic 2', 'Topic 3', 'Sum'])
    for ind,row in clusters_sum2[half].iterrows():
        if (row['Topic 1'] in group) or (row['Topic 2'] in group) or (row['Topic 3'] in group):
            pass
        else:
            clusters_sum3[half] = clusters_sum3[half].append(row, ignore_index=True)
            group.append(row['Topic 1'])
            group.append(row['Topic 2'])
            group.append(row['Topic 3'])

    line1 = []
    line2 = []
    line3 = []
    for ind,row in clusters_sum3[half].iterrows():
        line1.append(words[int(row['Topic 1'])-1])
        line2.append(words[int(row['Topic 2'])-1])
        line3.append(words[int(row['Topic 3'])-1])

    clusters_sum3[half]['Words 1'] = line1
    clusters_sum3[half]['Words 2'] = line2
    clusters_sum3[half]['Words 3'] = line3
```

```python
## Extract clusters
rel = [12]
sci = [7]
pol = [34]
cluster_rel = {}
cluster_sci = {}
cluster_pol = {}
for half in h_centuries2:
    cr = []
    cp = []
    cs = []
    for i in range (0,4):
        if (clusters_sum3[half].iloc[i,0] in rel) or (clusters_sum3[half].iloc[i,1] in rel) or (clusters_sum3[half].iloc[i,2] in rel):
            cr.append(clusters_sum3[half].iloc[i,0])
            cr.append(clusters_sum3[half].iloc[i,1])
            cr.append(clusters_sum3[half].iloc[i,2])
        elif (clusters_sum3[half].iloc[i,0] in sci) or (clusters_sum3[half].iloc[i,1] in sci) or (clusters_sum3[half].iloc[i,2] in sci):
            cs.append(clusters_sum3[half].iloc[i,0])
            cs.append(clusters_sum3[half].iloc[i,1])
            cs.append(clusters_sum3[half].iloc[i,2])
        elif (clusters_sum3[half].iloc[i,0] in pol) or (clusters_sum3[half].iloc[i,1] in pol) or (clusters_sum3[half].iloc[i,2] in pol):
            cp.append(clusters_sum3[half].iloc[i,0])
            cp.append(clusters_sum3[half].iloc[i,1])
            cp.append(clusters_sum3[half].iloc[i,2])
    cluster_rel[half] = cr
    cluster_sci[half] = cs
    cluster_pol[half] = cp
    
```
