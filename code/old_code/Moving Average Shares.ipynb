```python
df = cross.copy()
df = df.drop(columns = ['HTID', 'Year_rounded'])
tt = pd.DataFrame(df.columns, columns = ['Cross-Topic'])
tt
```

```python
breadth_by_year = {}
df = cross.copy()
df = df.drop(columns = 'HTID')
for year in years:
    b = df[(df['Year_rounded'] >= (year-10)) & (df['Year_rounded'] <= (year+10))]
    b = b.drop(columns = 'Year_rounded')
    ct = pd.DataFrame(b.columns, columns = ['Cross-Topic'])
    u = []
    for i in range(0,len(b.columns)):
        u.append(sum(b.iloc[:,i]))
    a = pd.concat([ct,pd.DataFrame(u)], axis=1)
    a.rename(columns={0:'breadth'}, inplace=True)
    breadth_by_year[year] = a
```

```python
breadth_by_year2 = {}
df = cross.copy()
df = df.drop(columns = 'HTID')
for year in years:
    b = df[(df['Year_rounded'] >= (year-10)) & (df['Year_rounded'] <= (year+10))].
    b = b.drop(columns = 'Year_rounded')
    ct = pd.DataFrame(df.columns, columns = ['Cross-Topic'])
    u = b.sum()
    a = pd.concat([ct,pd.DataFrame(u)], axis=1)
    a.rename(columns={0:'breadth'}, inplace=True)
    breadth_by_year[year] = a
```

```python
u = []
for year in years:
    u.append(breadth_by_year[year]['breadth'])
bct = pd.DataFrame(u)
bct = bct.transpose()
bct.columns = years
bct
```

```python
f = {}
cttot = []
for year in years: #Calculate shares
    total = sum(bct[year])
    cttot.append(total)
    norm = bct[year]/total
    f[year] = norm
    
    
ctshares = pd.DataFrame.from_dict(f)
ctshares 
```
